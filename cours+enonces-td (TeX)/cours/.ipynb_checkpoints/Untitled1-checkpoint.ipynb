```python
from itertools import *
import timeit
```

```python
couleurs = ["Pique", "Trefle", "Carreaux", "Coeur"]
valeurs = [2, 3, 4, 5, 6, 7, 8, 9, 10, "valet", "dame", "roi", "as"]

def f1():
    cartes = []
    for v in valeurs:
        for c in couleurs:
            cartes.append(str(v) + " de " + c)

%timeit f1()
```

```python
couleurs = ["Pique", "Trefle", "Carreaux", "Coeur"]
valeurs = [2, 3, 4, 5, 6, 7, 8, 9, 10, "valet", "dame", "roi", "as"]

def f2():
    cartes = []
    for v, c in product(valeurs, couleurs):
        cartes.append(str(v) + " de " + c)

%timeit f2()
```
