```python
def isFile(nom_fichier):
    try:
        f_log = open(nom_fichier, 'r')
        f_log.close()
        return(nom_fichier, True)
    except FileNotFoundError:
        return(nom_fichier, False)
```

```python
isFile("Ch11.pptx")
```

```python
isFile("aaa.pptx")
```

```python
# Programmme
while True:
    physique = input("nom du fichier: ")
    if physique == "":
        break
        # vérifier si le fichier existe
    if not isFile(physique)[1]:
        print("ce fichier n'existe pas")
    else:
        # afficher les lignes sur l'écran
        with open(physique, 'r') as f:
            for x in f:
                print(x)
```
